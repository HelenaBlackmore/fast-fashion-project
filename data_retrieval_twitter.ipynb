```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None) #enabling display of whole row in df
```

```python
# Creating list to append tweet data to
tweets_list_missguided = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(' content:"missguided"+OR+"@missguided"+OR+"#missguided" since:2018-05-03 until:2022-05-03 lang:"en" ').get_items()):

    tweets_list_missguided.append([tweet.date, tweet.id, tweet.content, tweet.renderedContent, tweet.replyCount, tweet.retweetCount, tweet.lang, tweet.source, tweet.media, tweet.url, tweet.user.username, tweet.user.id, tweet.user.friendsCount, tweet.user.location, tweet.place])
    
# Creating a dataframe from the tweets list above
tweets_df_missguided = pd.DataFrame(tweets_list_missguided, columns=['Datetime', 'ID', 'Content','Rendered Content','No. Replies','No. Retweets','Language','Source','Media','url','username', 'userid', 'friendcount', 'userlocation','tweetlocation'])
```

```python
#saving df in csv

tweets_df_missguided.to_csv('missguided_05.csv', encoding='utf16')
```
